there is a varying range which makes the machine learning model difficult to predict

let us take the data and fit it and then transform it using standarscaler funtion


```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data=scaler.transform(x)
```

now we need to train and split the data

here we use train variables to train the model and test variable to check the accuracy train and split function in used with x and y older dataset and test size as 20% i.e we use 20%of data to train the model

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,stratify=y,random_state=2) 
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

training the model

```python
classifer =svm.SVC(kernel ='linear') #svm used to load the svm model and svc is classifier and kernel is set to linear model
```

```python
# now we will feed the training data to this classifier
# training the support vector machine classifer
classifer.fit(x_train,y_train)
```

Model evaluation
evaluating: how many times our model is predicting correct values


accuracy score
now let us train the model using training data and cross check with the training data label and try to predict the accuracy score


```python
# accuracy score of training data
x_train_prediction=classifer.predict(x_train)  #now labels for xtrain is stored in this varible
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score of the training data:',training_data_accuracy)
```

```python
# now to checked the trained data let us check the accuray of the test data
x_test_prediction=classifer.predict(x_test)  
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score of the test data:',test_data_accuracy )
```
